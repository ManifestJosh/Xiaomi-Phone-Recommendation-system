```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('..//mi_data.csv')
```

```python
df
```

```python
from scipy.sparse import csr_matrix
```

```python
df['model_names']=df['model_names'].apply(lambda x:str(x.split('(')[0]))
df['Internal_Storage']= df['storage_ram'].apply(lambda x:str(x.split('RAM')[0]))
df['RAM']=df['storage_ram'].apply(lambda x:str(x.split('RAM')[1]))
df['OS']=df['os_processor'].apply(lambda x:str(x.split('Processor')[0]))
df['battery'] = df['battery'].apply(lambda x:str(x.replace('Battery Capacity','')))
    
```

```python
def network_type(x):
    if '5G' in x:
        return '5G'
    if '4G' in x:
        return '4G'
    else:
        return '3G'
```

```python
df['network'] = df['network'].apply(network_type)
df['Internal_Storage']=df['Internal_Storage'].str.replace('Internal Storage','') 
df['RAM']=df['RAM'].apply(lambda x:str(x.split('GB')[0]))
df['OS']=df['OS'].str.replace('Operating SystemAndroid','')
df['battery']=df['battery'].apply(lambda x:str(x.split('mAh')[0]))
```

```python
df.drop(['imgURL','storage_ram','os_processor'], axis=1, inplace=True)
```

```python
features_df=df.pivot_table(index='model_names', columns='Internal_Storage',values='ratings').fillna(0)
features_df.head(10)
```

```python
features_matrix= csr_matrix(features_df.values)
```

```python
features_df.shape
```

```python
from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(features_matrix)
```

```python
query_index = np.random.choice(features_df.shape[0])
query_index 
```

```python
model_knn.kneighbors
```

```python
distances, indices = model_knn.kneighbors(features_df.iloc[query_index,:].values.reshape(1,-1),n_neighbors=6)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendation for {0}:\n'.format(features_df.index[query_index]))
    else:
        print('{0}: {1},'.format(i, features_df.index[indices.flatten()[i]]))
```
